```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn.datasets import load_wine
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
wine = load_wine()
wine_data = wine.data
wine_target = wine.target
wine_data_normalized = StandardScaler().fit_transform(wine_data)
train, test, train_labels, test_labels = train_test_split(wine_data_normalized, wine_target, test_size=0.2, random_state=123)
```

```python
#Wizualization 2D
plt.scatter(train[:,0], train[:,1], c = train_labels, )
```

```python
#Wizualization 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(train[:,0], train[:,1], train[:,2] ,c = train_labels)
plt.show()
```

```python
from sklearn.metrics import pairwise_distances

class KNN_classyfier():
    '''
    Class that represent k nearest neighbours classifier.
    '''
    def __init__(self, n_neighbors):
        '''
        param: n_neighbors: number of neighbours.
        '''
        self.n_neighbors = n_neighbors
        
    def fit(self, train, train_target):
        '''
        param: train: training data.
        param: train_target: labels of training data. 
        '''
        self.train = train
        self.train_target = train_target
        
    def distances(self, data):
        '''
        Function that count distances between data and training data,
        using euclidean metric.
        param: data: vector or matrix to compute distances.
        '''
        if data.shape == (len(data),):
            return pairwise_distances([data], self.train)
        
        return pairwise_distances(data, self.train)
    
    def classify(self, data):
        '''
        Function that classify data to class based on the k nearest neighbours.
        param: data: data to classify.
        '''
        distance = self.distances(data)
        classes = []
        for index, dist in enumerate(distance):
            classes.append([])
            sorted_distances = sorted(dist)
            for k in range(self.n_neighbors):
                classes[index].append(self.train_target[np.where(sorted_distances[k] == dist)[0][0]])
        return classes
    
    def predict(self, data):
        '''
        Function that predicts, class of given data.
        param: data: data to predict.
        '''
        
        classes = self.classify(data)
        class_ = []
        for i in classes:
            counts = np.bincount(i)
            class_.append(np.argmax(counts))
        return np.array(class_)
    
    def score(self, test, labels):
        '''
        Function that compute accuracy score of given data.
        param: test: test data
        paramL labels: labels of test data.
        '''
        Z = self.predict(test)
        score = np.sum(Z == labels) / len(Z)
        return score
        
```

```python
train_score = []
test_score = []
for i in range(1,100):
    knn = KNN_classyfier(i)
    knn.fit(train, train_labels)
    train_score.append(knn.score(train, train_labels))
    test_score.append(knn.score(test, test_labels))
```

```python
#How accuracy changes with more neighbours.

plt.plot(range(1,100), train_score, 'r-')
plt.plot(range(1,100), test_score, 'b-')
plt.xlabel('k neighbours')
plt.ylabel('Accuracy')
plt.show()
```
