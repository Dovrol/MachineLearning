```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
X = 2 * np.random.rand(100, 1)             # Some random points
X_b = np.hstack((np.ones((len(X), 1)), X)) # Adding bias as 0 index
y = 4 + 3 * X + np.random.randn(100, 1)    # Labels
```

```python
plt.plot(X, y, 'b.')
plt.show()
```

$\theta$ = $(X^T*X)$*$X^T$*y <br />
Closed from solution for theta/weights

```python
#CLOSED FORM SOLUTION
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best # Ideal would be bias: 4 and weight 3.
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.hstack((np.ones((len(X_new), 1)), X_new))
y_predict = X_new_b.dot(theta_best)
```

```python
plt.plot(X_new, y_predict, 'r-')
plt.plot(X, y, 'b.')
plt.show()
```
