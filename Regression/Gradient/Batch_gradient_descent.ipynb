```python
import numpy as np
import matplotlib.pyplot as plt
X = 2 * np.random.rand(100, 1)             # Some random points
X_b = np.hstack((np.ones((len(X), 1)), X)) # Adding bias as 0 index
y = 4 + 3 * X + np.random.randn(100, 1)    # Labels

```

```python
class batchGD:
    def __init__(self, lr = 0.1, iter_ = 1000):
        self.lr = lr
        self.iter = iter_
        self.hist = []
        
    def fit(self, X, y):
        self.m = len(X)
        X_b = np.hstack((np.ones((len(X), 1)), X))
        self.weights = np.random.randn(2,1)
        for i in range(self.iter):
            gradients = 2/self.m * X_b.T.dot(X_b.dot(self.weights) - y)
            self.weights -= self.lr* gradients
            if i < 6:
                self.hist.append([list(self.weights[0]), list(self.weights[1])])
        return self.hist
    def predict(self, test):
        test_b = np.hstack((np.ones((len(test), 1)), test))
        return test_b.dot(self.weights)      
```

```python
bgd = batchGD()
hist = bgd.fit(X, y)
new_point = np.array([[0], [2]])
new_point_B = np.hstack((np.ones((len(new_point), 1)), new_point))
```

```python
plt.plot(X, y, 'b.')
for i in hist:
    predicted = new_point_B.dot(i)
    plt.plot(new_point, predicted, 'r-')
```
